# Numeric Type Conversion: Examine the Number of Installs

**Challenge**: How many apps had over 1 billion (that's right - BILLION) installations? How many apps just had a single install? 

Check the datatype of the Installs column.

Count the number of apps at each level of installations.

# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

**Challenge**: Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
df_apps['Price'] = pd.to_numeric(df_apps['Price'].apply(lambda x:x[1:])) # convert all string price into float
df_apps['Price'] = df_apps['Price'].fillna(0) #change all nan values in the oclumn into 0
df_apps['Price']
```

### The most expensive apps sub $250

```python
df_apps = df_apps[df_apps['Price']<=250] # Remove all apps > $250
df_apps.sort_values('Price', ascending=False)
```

### Highest Grossing Paid Apps (ballpark estimate)

```python
revenue_estimate = df_apps['Price'] * df_apps['Installs']
df_apps.insert(1, 'Revenue_Estimate', revenue_estimate) # Insert Revenue_Estimate Column
df_apps.sample(5)
```

```python
highest_grossing_paid_apps_df = df_apps.sort_values('Revenue_Estimate', ascending=False, ignore_index=True)
highest_grossing_paid_apps_df.head(10) # 7 of them are basically games
```

# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? # Grouped Bar Charts: Free vs. Paid Apps per Category

```python
# free_apps = df_apps[df_apps['Type']=='Free']['Category'].value_counts()
# paid_apps = df_apps[df_apps['Type']=='Paid']['Category'].value_counts()
# free_vs_paid_df = pd.merge(free_apps, paid_apps, right_index=True, left_index=True)
# free_vs_paid_df.rename(columns={'Category_x':'Free', 'Category_y':'Paid'}, inplace=True)
# free_vs_paid_df

df_free_vs_paid = df_apps.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
```
